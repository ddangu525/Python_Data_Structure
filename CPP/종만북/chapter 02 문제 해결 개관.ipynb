## 종만북 chapter 2. 문제 해결 개관
- 문제를 푸는 기술 연마
    - 자신이 문제를 어떤 방식으로 해결하는지 의식
        - 부족한 부분, 개선점 파악

### 문제 해결 과정

#### 파인만 알고리즘
1. 문제를 적는다.
2. 골똘히 생각한다.
3. 답을 적는다.

문제 해결 과정을 단계별로 나누었다는 게 중요하다!!  
문제 해결을 위해 거쳐야 하는 과정들을 세분화하여 어디가 부족하고 어디를 개선해야 하는지 판단할 수 있다.  
   
How to Solve It에서는...  
1. 문제를 '이해'한다.
2. 어떻게 풀지 '계획'을 세운다.
3. 계획을 '수행'하여 문제를 해결한다.
4. 어떻게 풀었는지 돌아보고, 개선할 방법이 있는지 찾아본다.
  
우리는 6단계로 만들어보자.
1. 문제를 읽고 이해한다.
    - 제약조건을 빼먹지 않도록 주의.
    - 실수하지 않도록 정확하게..
2. 문제를 익숙한 용어로 재정의한다.(재정의와 추상화)
    - 자신이 다루기 쉬운 개념을 이용하여 문제를 자신의 언어로 풀어쓰기.
    - 문제가 요구하는 바를 직관적으로 이해, 복잡한 문제일수록 더 중요.
    - 어떤 부분을 추상화할 것인지를 선택.
3. 어떻게 해결할지 계획을 세운다.
    - 문제를 어떤 방식으로 해결할지 결정.
    - 사용할 알고리즘과 자료구조를 선택.
    - 가장 중요한 단계
4. 계획을 검증한다.
    - 설계한 알고리즘이 요구 죠건을 정확히 수행하는지 증명
    - 시간과 메모리가 문제의 제한을 초과하지 않는지 확인.
5. 프로그램으로 구현한다.
    - 풀이를 코드로 구현.
6. 어떻게 풀었는지 돌아보고, 개선할 방법이 있는지 찾아본다.(회고)
    - 장기적으로 가장 큰 영향!!
    - 문제를 한 번만 풀어서는 모든 걸 다 배우지 못한다.
    - 코드와 함께 자신의 경험을 기록으로 남기자.
        - 간단한 해법
        - 접근법
        - 깨달음
        - 오답원인
               - 자주 틀리는 부분..
               - 자주하는 실수
    - 다른 사람의 코드 보기.

- 문제를 풀지 못하여 다른사람의 코드를 본다면 반드시 복기하기!
    

### 문제 해결 전략
어려운 문제일수록 다양한 방법을 시도해 보면서 답안을 찾자.

#### 직관과 체계적인 접근
- 직관: 문제 해결에 필요한 알고리즘이 대략적으로 어떤 형태를 가질지를 짐작.
- 체계적인 접근: 바로 풀이법이 떠오르지 않는 어렵고 막막한 문제.
  
**체계적인 접근을 위한 질문들**
1. 비슷한 문제를 풀었었나?
    - 철도망 위에서 두 도시를 잇는 가장 짧은 경로를 찾는 문제를 풀었다면.
        - 한 도시를 두 번 방문하지 않으면서 가장 긴 경로를 찾는 문제
        - 기차를 네 번 이하로 갈아타면서 가장 짧은 경로를 찾는 문제
        - 역 간 운행 거리 중 가장 긴 구간이 가장 짧은 경로를 찾는 문제
        - 역 간 운행 거리 중 가장 짧은 구간이 가장 긴 경로를 찾는 문제
        - 가장 긴 구간과 가장 짧은 구간의 길이 차이가 가장 적은 경로를 찾는 문제..
    응용해서 풀 수 있는 것도 있지만 없는 것도 있음.  
    구분하려면 최단 경로 알고리즘을 단순히 알고 있는 것에 그치지 않고 동작 과정과 원리를 완전히 이해해야 한다.
    - 어떤 사건의 발생 확률이나 경우의 수 계산.
        - 대부분 동적 계획법으로 해결 가능.
2. 단순한 방법에서 시작할 수 있을까?
    - 시간과 공간 제약을 생각하지 않고 무식하게 풀기. 가장 단순한 알고리즘을 짜보는 것.
    - 간단하게 풀 수 있는 문제를 너무 복잡하게 생각해서 어렵게 푸는 실수를 예방.
3. 사탕 분배 문제 case study.
    - N(N<=30)개의 사탕을 세 명의 어린이에게 가능한 공평하게 나눠주려고 한다. (받은 사탕의 무게가 차이가 적게 나도록). 사탕의 무게는 모두 20 이하의 정수.. 가능한 최소 차이는?
        1. 가장 단순한 방법: 사탕을 나눠주는 모든 방법을 다 해보기.. 각 사탕마다 어느 어린이에게 줄 지 결정한다면 $3^N$, 최대 205조개..
        2. 위에서 중복으로 세는 경우를 없애주자.
            1. 실제로 받은 사탕이 다르더라도 무게가 같다면 문제의 값은 같다.
                - 어린이가 가진 사탕의 총량을 상태 공간으로 하는 BFS도 가능.
                    - 각 상태를 세 어린이가 가진 사탕의 양을 표현하는 세 개의 정수. (0, 0, 0) ~
                    - 사탕의 최대 무게는 20이므로 각 어린이가 가진 사탕의 양은 0 ~ 20xN.
                    - 각 상태에 대해 방문 여부를 저장하려면 크기가 $(20*N+1)^3$ 최대 2억 크기의 배열 필요.
            2. 사탕의 최대 무게는 20. 차이가 20 이상이 난다면 많이 가진 애가 적게 가진애에게 주면 차이는 항상 줄어듦..
                - 무게 차이가 20 이상인 경우 모두 무시 가능.. 실제로 할당해야 하는 배열 크기는 대략 1000만으로 줄어든다.
            3. 3명 중 누가 많이받고, 적게받는지는 중요하지 않음.. 항상 정렬된 경우만 탐색. 다시 1/6 줄어듦.
        3. 최종적으로 대략 200만개.. 처음의 1억분의 1정도로 검사해야할 상태의 수가 줄어든다....
        4. 동적 계획법이 이러한 맥락.. 
4. 내가 문제를 푸는 과정을 수식화할 수 있을까?
    - 손으로 여러 간단한 입력, 예제입력을 직접 해결해 보기.
5. 문제를 단순화할 수 없을까?
    - 주어진 문제의 좀더 쉬운 변형판을 먼저 풀어 보는 것. 
        - 제약 조건을 없애거나, 변수의 수 줄이기, 다차원 문제를 1차원으로 줄여서 표현..
        - 2차원 격자 위에 N개의 점이 있을 때, 이들까지의 맨하튼 거리의 합이 최소가 되는 새 점의위치??
            - 주어진 그대로 푸는 대신 1차원 형태로 바꿔 풀 수 있다.
            - 어떤 후보 x의 왼쪽에 있는 점들이 오른쪽에 있는 점들보다 많다면 x를 좀더 왼쪽으로 옮기는 것이 이득.
            - x축, y축에 각각 사영하여 2개의 작은 문제를 만들고 합하면 원 문제의 답과 같다.
6. 그림으로 그려볼 수 있을까?
    - 문제에 관련된 그림을 그려 보면 직관적 이해를 얻을 수도 있다. 
7. 수식으로 표현할 수 있을까?
    - 평문으로 쓰여 있는 문제를 수식으로 표현하는 것도 직관을 얻기 좋다.
    - 순수한 수학적 조작이 문제를 해결하는 데 큰 도움이 될 수도. 
8. 문제를 분해할 수 있을까?
    - 문제를 더 다루기 쉬운 형태로 변형하는 것.
    - 제약 조건 분해
        - 주어진 복잡한 조건을 단순한 형태를 갖는 조건의 집합으로 분해.
9. 뒤에서부터 생각해서 문제를 풀 수 있을까?
    - 문제에 내재된 순서를 바꿔보기.
    - 사다리게임 문제. 
        - 어디를 선택해야 좋은 결과를 얻을 수 있을지 바로 알려면 거꾸로 올라가보면 된다.
    - 삽입 정렬 뒤집기 / 감시카메라 설치 / Sorting Game...
10. 순서를 강제할 수 있을까?
    - 순서를 뒤집는 방법의 연장선. 순서가 없는 문제에 순서를 강제
    - 불켜기 문제.
    - 경우의 수를 셀 때도 유용. 
11. 특정 형태의 답만을 고려할 수 있을까?
    - 순서 강제의 연장선으로 정규화(canonicalization) 기법이 있다.
    - 우리가 고려해야 할 답들 중 형태가 다르지만 결과적으로 같은 것들을 그룹으로 묶은 뒤, 각 그룹의 대표들만 고려하는 방법.
    - 원 덮기 문제...
        - 무한개의 후보 대신 유한한 부분 집합만을 고려.
    - 매우 유용하지만 기법이 문제마다 다르기 때문에 많은 경험 필요...
    - 소풍 / 단어 제한 끝말잇기 문제..
