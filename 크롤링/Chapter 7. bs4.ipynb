## 1. 파서의 차이
사용가능한 파서로 lxml, html5lib, html.parser 세 가지.. 
### 1-1 lxml
c언어로 구현. 가장 빠름. XML도 처리 가능. c언어에 의존적.

```python
from bs4 import BeautifulSoup as bs

html = """<p>test</p>"""

soup = bs(html, 'lxml')
print(soup)
```

### 1-3 lxml VS html5lib
속도차이때문에 lxml을 사용해야 한다.

```python
# 속도 측정코드
import time
startTime = time.time()

print('측정하고 싶은 실행 코드')

endTime = time.time() - startTime
print(endTime)
```

```python
html = """<html><head></head><p>test</p></html>"""

start_time = time.time()
bs(html, 'lxml')
lxml_end_time = time.time() - start_time

start_time = time.time()
bs(html, 'html5lib')
html5lib_end_time = time.time() - start_time

print(f'lxml 시간측정 : {lxml_end_time}\nhtml5lib 시간측정 : {html5lib_end_time}\n{html5lib_end_time/lxml_end_time}')
```

```python
# 반복문으로 어느정도 차이나는지 평균값 확인

time_sum = 0
loop_count = 100
for i in range(loop_count):
    
    start_time = time.time()
    bs(html, 'lxml')
    lxml_end_time = time.time() - start_time
    
    start_time = time.time()
    bs(html, 'html5lib')
    html5lib_end_time = time.time() - start_time
    
    rate = html5lib_end_time / lxml_end_time
    
    print(f'{i+1}번째 시도\nlxml 시간측정: {lxml_end_time}\nhtml5lib 시간측정: {html5lib_end_time}\nrate: {rate}\n')
    time_sum += rate
average = time_sum / loop_count
print(f'평균비율 : {average}')
```

### 2-5 부모 태그 접근하기
자식태그로 내려가는 형태가 아니라 부모 태그로 올라가는 형태로도 접근 가능.

```python
# parent
tag_span = soup.span
tag_title = soup.title

span_parent = tag_span.parent
title_parent = tag_title.parent

print(f'태그\n{tag_span}\n{tag_title}\n부모태그\n{span_parent}\n{title_parent}')
```

parents는 children과 비슷하지만 제너레이터로 가져옴.. 이것도 반복문 이용.

```python
print('span 부모 태그')
for parent in span_parents:
    print(parent)
print('title 부모 태그')
for parent in title_parents:
    print(parent)
```

parents 속성을 이용하면 가장 최상위 부모 태그까지!

### 2-6 형제 태그 접근하기
형제(sibling)태그란 동등한 위치의 태그를 의미. parent가 같은.. 태그가 같을 필요는 없다.

```python
# 형제 태그 접근
a = tag_span.next_sibling
b = a.previous_sibling
print(a)
print(b)
```

next_sibling, previous_sibling으로 다음형제와 이전 형제를 가져올 수 있다.

```python
# 형제 태그 접근 2
html = """<html> <head><title>teest site</title></head> <body><p><a>test1</a><b>test2</b><c>test3</c></p> </body></html>"""
soup = bs(html, 'lxml')

tag_a = soup.a
tag_b = soup.b
tag_c = soup.c

print(f'tags\n{tag_a}\n{tag_b}\n{tag_c}\n')

tag_a_next = tag_a.next_sibling
tag_a_prev = tag_a.previous_sibling

tag_b_next = tag_b.next_sibling
tag_b_prev = tag_b.previous_sibling

tag_c_next = tag_c.next_sibling
tag_c_prev = tag_c.previous_sibling

print(f'siblings\n\n{tag_a_next}\n{tag_a_prev}\n')
print(f'{tag_b_next}\n{tag_b_prev}\n')
print(f'{tag_c_next}\n{tag_c_prev}\n')
```

가장 처음에 나오는 에는 previous_sibling하면 None값.. 가장 마지막의 next_sibling도 None.  
얘네들도 복수형태로 s 붙여서 사용 가능.

```python
# 형제 태그 접근 3

tag_a_nexts = tag_a.next_siblings
tag_a_prevs = tag_a.previous_siblings

print(tag_a_nexts)
print(tag_a_prevs)

tag_b_nexts = tag_b.next_siblings
tag_b_prevs = tag_b.previous_siblings

print(tag_b_nexts)
print(tag_b_prevs)

tag_c_nexts = tag_c.next_siblings
tag_c_prevs = tag_c.previous_siblings

print(tag_c_nexts)
print(tag_c_prevs)
```

얘네도 generator. 반복문으로 보자

```python
print('next siblings')
for sibling in tag_a_nexts:
    print(sibling)
print('previous siblings')
for sibling in tag_a_prevs:
    print(sibling)
```

### find_all()
find_all() 함수를 이용하면 우리가 원하는 태그들을 리스트의 형태로 얻어올 수 있다.

```python
# 원하는 태그 전부 가져오기
html = """<html> <head><title>test site</title></head> <body><p>test1</p><p>test2</p><p>test3</p> </body></html>"""

soup = bs(html, 'lxml')

print(soup.find_all('title'))
print(soup.find_all('p'))
```

soup.p를 했을 때는 가장 처음 등장하는 p 태그만 가져왔지만, find_all() 함수를 이용하면 첫 번째 인자로 전달된 태그를 전부 가져와서 리스트로..

```python
# id 값으로 가져오기. 하나만 존재하거나 빈 리스트..
html = """<html> <head><title>test site</title></head> <body><p>test1</p><p id="d">test2</p><p>test3</p> </body></html>"""
soup = bs(html, 'lxml')
print(soup.find_all(id='d'))
```

```python
# id 존재 유무로 가져오기
print(soup.find_all(id=True))
print(soup.body.find_all(id=False))
```

soup는 문서 전체를 의미. HTML안에 있는 모든 것을 기준으로 출력. 필요없는 태그들도 반환됨..

### 2-13 extract()
extract()는 태그를 지우는 역할. 실제 웹 사이트에서 HTML을 가져오면 용량이 크기 때문에 style, script같이 크롤링에 필요없는 태그는 제거

```python
# 필요 없는 태그 제거 1
a = soup.body.extract()

print('제거항목')
print(a)
print('제거완료')
print(soup)
```

soup.tag.extract()로 해당 태그 제거.  
body를 제거한 것. 실제로는 body만 필요!

```python
# 필요 없는 태그 제거 2
html = """<html> <head><title>test site</title></head>
<body><div><p id="i" class="a">test1</p><p class="d">test2</p></div><p class="d">test3</p> <a>a tag</a> <b> b tag</b> </body></html>"""
soup = bs(html, 'lxml')

for tag in soup.select('p'):
    print(tag.extract())
print('제거완료')
print(soup)
```

find_all(), find(), select()를 이용하여 요소에 접근한 후 extract()를 이용하여 해당 태그를 제거.

```python
# 필요 없는 태그 제거 함수
def remove_tag(soup, tags):
    if tags == []:
        return False
    removes = []
    
    for tag in soup.find_all(tags):
        removes.append(tag.extract())
    return removes

html = """<html> <head><title>test site</title></head> \
        <body><div><p id="i" class="a">test1</p><p class="d">test2</p></div>\
        <p class="d">test3</p> <a>a tag</a> <b> b tag</b> </body></html>"""
soup = bs(html, 'lxml')

removed_tag = remove_tag(soup, 'p')
print(removed_tag)
print(soup)
```

```python
html = """<html> <head><title>test site</title></head> <body>\
        <div><p id="i" class="a">test1</p><p class="d">test2</p></div> \
        <p class="d">test3</p> <a href="/example/test1">a tag</a> <b> b tag</b> </body></html>"""
soup = bs(html, 'lxml')

print(soup.find_all(class_=re.compile('d')))   # 클래스 값에 d가 포함된 요소를 찾음
print(soup.find_all(id=re.compile('i')))       # 아이디 값에 i가 포함된 요소를 찾음
print(soup.find_all(re.compile('t')))          # 태그에 t가 포함된 요소
print(soup.find_all(re.compile('^t')))         # 태그 이름이 t로 시작하는 요소(^는 시작을 의미)
print(soup.find_all(href=re.compile('/')))     # href에 슬래시(/)가 포함된 요소
```

### 3-2 정규식
정규식 사용을 위해 세 가지 과정을 거침
1. re.compile(정규 표현식)을 이용하여 패턴을 만든다.
2. 만들어진 패턴을 이용해서 match(문자열), search(문자열), findall(문자열), finditer(문자열)을 한다.
3. 2단계에서 match(), search()를 통해 나온 결과물을 group(), start(), end(), span()을 이용하여 리턴한다.  
findall()과 finditer()은 리스트와 객체로 반환하므로 반복문을 이용하여 group(), start(), end(), span() 사용해야..

|Method|목적|
|---|---|
|match()|문자열의 처음부터 정규식과 매치되는지 조사한다.|
|search()|문자열 전체를 검색하여 정규식과 매치되는지 조사한다.|
|findall()|정규식과 매치되는 모든 문자열(substring)을 리스트로 리턴한다|
|finditer()|정규식과 매치되는 모든 문자열(substring)을 iterator 객체로 리턴한다|

```python
# 정규 표현식 사용
test_str = "test t1sd j test1"

pattern = re.compile('test')
a = pattern.match(test_str)
b = pattern.search(test_str)
c = pattern.findall(test_str)
d = pattern.finditer(test_str)

print('-- match result --')
print(a)
print(a.group(), a.start(), a.end(), a.span())

print('-- sarch result --')
print(b)
print(b.group(), b.start(), b.end(), b.span())

print('-- findall result --')
print(c)

print('-- finditer result --')
print(d)
for i in d:
    print(i.group(), i.start(), i.end(), i.span())
```

```python
# 간단한 크롤러 만들기 2
import requests as rq
from bs4 import BeautifulSoup as bs

base_url = "https://pjt3591oo.github.io/"
page = 2
page_path = f'/page{page}'

res = rq.get(base_url)
soup = bs(res.content, 'lxml')

posts = soup.select('body main.page-content div.wrapper div.home div.p')

for post in posts:
    title = post.find('h3').text.strip()
    descript = post.find('h4').text.strip()
    author = post.find('span').text.strip()
    print(title, descript, author)

while True:
    sub_path = page_path
    page += 1
    res = rq.get(base_url + sub_path)
    
    if res.status_code != 200:
        break
        
    soup = bs(res.content, 'lxml')
    
    posts = soup.select('body main.page-content div.wrapper div.home div.p')
    
    for post in posts:
        title = post.find('h3').text.strip()
        descript = post.find('h4').text.strip()
        author = post.find('span').text.strip()
        print(title, descript, author)
```
