## 1. 돔 접근
크롤러를 만드는 데 돔 접근이 가장 중요.  
데이터 수집뿐 아니라 마우스, 키보드 관련 이벤트도 발생시킬 수 있다!!  
selenium으로 돔 접근을 할 경우 bs4와 약간 차이가 있음.  
selenium에서 제공하는 돔 접근 함수는 bs4보다 많다.  

---
1. **반환 타입: 단일 객체**(bs4에서 find()같은 형태로 반환)  
find_element_by_id  
find_element_by_tag_name  
find_element_by_xpath  
find_element_by_link_text  
find_element_by_partial_link_text  
find_element_by_name  
find_element_by_class_name  
find_element_by_css_selector  
2. **반환타입: 리스트**(bs4에서 find_all()같은 형태로 반환)  
find_elements_by_tag_name  
find_elements_by_xpath  
find_elements_by_link_text  
find_elements_by_partial_link_text  
find_elements_by_name  
find_elements_by_class_name  
find_elements_by_css_selector  

bs4에서는 find()함수나 find_all()함수에 클래스, 아이디, 속성, 태그 이름 인자를 전달해 줌으로써 돔 접근.  
셀레니움에서는 인자가 아니라 각각 함수로 제공.  
id 제외한 다른 방법은 모두 리스트 형태나 단일객체 두 형태로 제공된다.   
xpath는 XML 문서에서 접근하기 위한 문법.  
link_text, partial_link_text는 a태그의 href 속성값을 이용하여 돔을 찾는 방법.

```python
# selenium 웹 브라우저 띄우기
from selenium import webdriver
driver = webdriver.Chrome('chromedriver.exe')
```

반환된 객체는 웹 브라우저를 의미한다. 이 브라우저 객체를 driver 변수에 저장. 이걸 사용해서 웹 조작.  
**얘가 켜져있어야 계속 조작 가능하다!!!**

```python
# 웹 접속하기
url = 'https://pjt3591oo.github.io'
driver = webdriver.Chrome('chromedriver.exe')
driver.get(url)
```

웹 브라우저를 띄운 후 크롤러를 만들기 위한 사이트에 접속해줌.  
get() 함수를 이용하여 사이트 접속. 위 코드를 실행시키면 빈 웹 브라우저가 뜬 다음 URL 변수에 저장된 주소로 접속함.

### 1-1 find_element_by_id
id 값을 이용하여 돔 접근하는 함수.

```python
selected_id = driver.find_element_by_id('nav-trigger')
print(selected_id)
print(selected_id.tag_name)
print(selected_id.text)
```

### 1-2 find_element_by_tag_name / find_elements_by_tag_name
태그 이름을 이용한 돔 접근. 하나 또는 리스트.

```python
# 태그 이용하여 접근
selected_tag_p = driver.find_element_by_tag_name('p')
print(selected_tag_p.tag_name)
print(selected_tag_p.text)

selected_tags_p = driver.find_elements_by_tag_name('p')
print(selected_tags_p)
```

elements로 리스트를 반환할 때는 돔의 세션을 리스트로 만들어 반환해줌.  
각각 tag_name이나 text 속성에 접근하고 싶다면 반복문 이용.

### 1-3 find_element_by_name / find_elements_by_name
name 속성값을 찾아주는 함수. name 속성은 input 태그에 많이 사용됨.

```python
# name 속성 이용하여 접근
url = 'https://pjt3591oo.github.io/search'

driver = webdriver.Chrome('chromedriver.exe')
driver.get(url)

selected_name = driver.find_element_by_name('query')
print(selected_name)
print(selected_name.tag_name)
print(selected_name.text)

selected_names = driver.find_elements_by_name('query')
print(selected_names)
```

### 1-4 find_element_by_link_text / find_elements_by_link_text
a 태그의 href를 이용하여 수집.  
URL 모든 요소를 검사하는 게 아니라 쿼리스트링 부분만 찾는다

```python
# href 속성 이용하여 접근 - 쿼리스트링
url = 'https://pjt3591oo.github.io'
driver = webdriver.Chrome('chromedriver.exe')
driver.get(url)

selected_link = driver.find_element_by_link_text('')
print(selected_link)
print(selected_link.tag_name)
print(selected_link.text)

selected_links = driver.find_elements_by_link_text('')
print(selected_links)
```

### 1-5 find_element_by_partial_link_text / find_elements_by_partial_link_text
쿼리스트링이 아니라 path를 이용하여 찾는 방법

```python
# href속성 이용하여 접근 - path - 1
# href속성에 path를 의미하는 부분에 react가 포함된 태그.
selected_link = driver.find_element_by_partial_link_text('react')
print(selected_link)
print(selected_link.tag_name)
print(selected_link.text)

selected_links = driver.find_elements_by_partial_link_text('react')
print(selected_links)
```

```python
# href속성 이용하여 접근 - path - 2
# 빈 값을 넣는다면 href속성이 있는 태그.
selected_link = driver.find_element_by_partial_link_text('')
print(selected_link)
print(selected_link.tag_name)
print(selected_link.text)

selected_links = driver.find_elements_by_partial_link_text('')
print(selected_links)
```

### 1-6 find_element_by_class_name / find_elements_by_class_name
클래스를 이용해 찾기

```python
# 클래스 이용하여 접근
# 클래스가 p인 모든 태그 찾기
selected_class = driver.find_element_by_class_name('p')
print(selected_class)
print(selected_class.tag_name)
print(selected_class.text)

selected_classes = driver.find_elements_by_class_name('p')
print(selected_classes)
```

### 1-7 find_element_by_css_selector / find_elements_by_css_selector
bs4에서 select()처럼 CSS셀렉터를 이용하여 돔 접근. bs4에서는 무조건 리스트 형태였지만 selenium에서는 두 가지 타입 모두 반환 가능

```python
# 셀렉터 이용하여 접근
selected_selector = driver.find_element_by_css_selector('div.home div.p')
print(selected_selector)
print(selected_selector.tag_name)
print(selected_selector.text)

selected_selectors = driver.find_elements_by_css_selector('div.home div.p')
print(selected_selectors)
```

### 1-8 없는 요소 접근
bs4에서와 다르게 없는 요소에 접근하면 에러를 띄움.

```python
# 없는 요소 접근
selected_selector = driver.find_element_by_css_selector('.a')
print(selected_selector)
print(selected_selector.tag_name)
print(selected_selector.text)
```

## 2. 웹 제어하기
selenium에서 돔에 접근하는 이유는 데이터 수집 목적보다 웹 제어 목적이 더 크다.  
웹 제어 종류로는 마우스 제어, 키보드 제어, JavaScript 삽입이 있다.

### 2-1 마우스 제어
마우스를 제어하기 위해 돔에 접근한 이후 click()함수를 호출

```python
# 마우스 클릭하기
selected_selector = driver.find_element_by_css_selector('div.home div.p a')
print(selected_selector)
print(selected_selector.tag_name)
print(selected_selector.text)
selected_selector.click()
```

click() 함수를 이용하면 해당 돔을 클릭한다. 하지만 이를 이용하여 페이지를 이동하는 것은 피하는 것이 좋다.

```python
# click 이용하여 페이지 이동
url = 'https://pjt3591oo.github.io'
driver.get(url)

selected_tags_a = driver.find_elements_by_css_selector('a')

for i in selected_tags_a:
    print(i.text, i.tag_name)
    i.click()
```

0번째 요소만 클릭해서 이동하고 인덱스 에러!!  
메인 페이지에서 돔을 가져오고 다른 페이지로 넘어가면 메인 페이지에서 가져온 돔에 접근할 수 없기 때문!!!  
click은 페이지 이동 용도로 쓰면 안된다!  
클릭했을 때 페이지 변화가 없고 해당 페이지 내에서 데이터가 추가되는 경우에 사용하는 것이 좋다.

```python
# 클릭을 사용하기 좋은 예
url = 'https://pjt3591oo.github.io/search'
driver.get(url)

selected_tag_a = driver.find_element_by_css_selector('input#search-box')

selected_tag_a.click()
```

### 3. selenium과 bs4의 조합
selenium 단독사용보다 bs4와 같이 사용하면 최고 효율.  
selenium을 사용하면 웹 브라우저를 띄워주기 위해 더 많은 자원을 사용하기 때문에 속도가 아주 느려짐.  

```python
# selenium VS bs4

# selenium 시간 측정
import time
url = 'https://pjt3591oo.github.io'

selenium_start = time.time()

driver = webdriver.Chrome('chromedriver.exe')
driver.get(url)

div_selectors11 = driver.find_elements_by_css_selector('div')
div_selectors12 = driver.find_elements_by_css_selector('div')
div_selectors13 = driver.find_elements_by_css_selector('div')
div_selectors14 = driver.find_elements_by_css_selector('div')
div_selectors15 = driver.find_elements_by_css_selector('div')

selenium_end = time.time() - selenium_start

print(f'usage selenium : {selenium_end}')

# bs4 사용시간 측정
import requests as rq
from bs4 import BeautifulSoup as bs

bs4_start = time.time()

res = rq.get(url)
soup = bs(res.content, 'lxml')
div_selectors21 = soup.select('div')
div_selectors22 = soup.select('div')
div_selectors23 = soup.select('div')
div_selectors24 = soup.select('div')
div_selectors25 = soup.select('div')

bs4_end = time.time() - bs4_start

print(f'usage bs4 : {bs4_end}')
```

10배 이상 차이.  selenium으로는 requests로 처리하기 힘든 JavaScript를 처리하고, 만들어진 HTML은 bs4를 이용하여 처리하자.

```python
# selenium으로 HTML 가져오기
driver.get(url)

print(driver.page_source)
```

웹 브라우저 객체인 driver에서 page_source 속성을 이용하여 현재 웹 브라우저의 HTML 코드를 가져옴.  
이 때, requests 모듈과 다르게 **JavaScript가 처리된 HTML코드**를 가져온다!!

```python
# selenium과 bs4 조합

driver.get(url)
soup = bs(driver.page_source)

print(soup.select('div'))
```

## 4. selenium 활용
selenium을 활용해보자. JavaScript를 이용하여 HTML을 추가하는 대표적인 예가 검색.  

```python
# 검색 결과 데이터 수집
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs

url = 'https://pjt3591oo.github.io/search'
search_keysword = 'db'

driver = webdriver.Chrome('chromedriver.exe')
driver.get(url)

selected_tag_a = driver.find_element_by_css_selector('input#search-box')

selected_tag_a.send_keys(search_keysword)
selected_tag_a.send_keys(Keys.ENTER)   # '\ue007'로 해도 됨.

soup = bs(driver.page_source, 'lxml')
items = soup.select('ul#search-results li')

for item in items:
    title = item.find('h3').text
    description = item.find('p').text
    print(title)
    print(description)
```
