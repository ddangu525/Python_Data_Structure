씨앗은 단 3개!  꽃 하나당 5개의 칸을 차지하는데 대여비용을 최소로 해야함. 겹치거나 밖으로 나가지 않게 해서 꽃 3개가 모두 만개하도록!  
그럼 일단 심는 지점은 (1,1)~(N-2,N-2) 사이가 되겠고... 최대 크기일 때 10x10 크기..  
9x9에서 3개를 선택? $_{81}C_3$ = 85320.  그리고 겹치는 경우 안됨.  이렇게 하면 되겠다.  

```python
# import sys
# input = sys.stdin.readline
from itertools import combinations
N = int(input())
price = [list(map(int, input().split())) for _ in range(N)]
num_coords = {(N-2)*(y-1)+x:(x,y) for y in range(1,N-1) for x in range(1,N-1)}  # x,y순서로 저장.
# 최소값 초기화
min_value = float('inf')
for c1, c2, c3 in combinations(range(1,(N-1)**2+1), 3):   # 전체 중에 3개 선택.
    x1, y1 = num_coords[c1]
    x2, y2 = num_coords[c2]
    x3, y3 = num_coords[c3]
    # 겹치는 경우 그냥 넘어가기
    if abs(x1-x2)+abs(y1-y2)<=2 or abs(x1-x3)+abs(y1-y3)<=2 or abs(x2-x3)+abs(y2-y3)<=2:
        continue
    value = price[y1][x1]+price[y2][x2]+price[y3][x3]
    if value<min_value:
        min_value = value
print(min_value)
```

```python
# import sys
# input = sys.stdin.readline
from collections import deque

N, M = map(int, input().split())
maze = [list(map(int, input().split())) for _ in range(N)]
# red인 경우 안가게.. 그리고 pink, orange는 한번씩만 방문..
visit = [[False]*M for _ in range(N)]

q = deque()
# x, y, d, smell orange, is before purple 총 5개 정보.
q.append((0,0,1,0,0))   # 아래쪽으로출발
q.append((0,0,3,0,0))   # 오른쪽으로출발
visit[0][0] = True
move = 1
while q:
    # movement별로 찾기 위해.
    for _ in range(len(q)):
        cx, cy, cd, o, purple = q.popleft()
        
        # 일단 방문여부부터. 
        # 1. 얘는 왼쪽으로 가는경우.. x가 작아짐. 이 경우에는 답이 나올 수가 없다.
        if cx>0 and not visit[cy][cx-1]:
            # red로 못감
            if maze[cy][cx-1]==0:
                visit[cy][cx-1] = True
            # pink는 그대로 진행
            elif maze[cy][cx-1]==1:
                visit[cy][cx-1] = True
                q.append((cx-1,cy,o))
            # orange인경우 오렌지냄새가 생김! 방문체크 해줘야하나...
            elif maze[cy][cx-1]==2:
                visit[cy][cx-1] = True
                q.append((cx-1,cy,1))
            # blue인경우 오렌지향이 있어야만 가능
            elif maze[cy][cx-1]==3 and o==1:
                visit[cy][cx-1] = True
                q.append((cx-1,cy,1))
            # !!!!! purple인경우. 그 방향으로 nonpurple이 나오거나, 못가는 칸이 나올때까지 진행.냄새도 없어짐.
            else:
                for x in range(cx-2,-1,-1):
                    # 이동못하면 이전칸에 멈추기.
                    if maze[cy][x]==0 or (maze[cy][x]==3 and o==0):
                        q.append((x+1,cy, 0))
                        break           
                    # 이동가능하고 nonpurple인경우 그 칸에 멈추기
                    elif maze[cy][x]==1 or maze[cy][x]==2 or (maze[cy][x]==3 and o==1) or x==0:
                        q.append((x,cy,0))
                        break
        
        # 2. 오른쪽으로 가는 경우. 이 경우엔 답 나올 수 있음..
        if cx<M-1 and not visit[cy][cx+1]:
            # 답인경우!
            if (cx+1, cy)==(M-1,N-1):
                print(move)
                exit()
            # red로 못감
            if maze[cy][cx+1]==0:
                visit[cy][cx+1] = True
            # pink는 그대로 진행
            elif maze[cy][cx+1]==1:
                visit[cy][cx+1] = True
                q.append((cx+1,cy,o))
            # orange인경우 오렌지냄새가 생김! 방문체크 해줘야하나...
            elif maze[cy][cx+1]==2:
                visit[cy][cx+1] = True
                q.append((cx+1,cy,1))
            # blue인경우 오렌지향이 있어야만 가능
            elif maze[cy][cx+1]==3 and o==1:
                visit[cy][cx+1] = True
                q.append((cx+1,cy,1))
            # !!!!! purple인경우. 그 방향으로 nonpurple이 나오거나, 못가는 칸이 나올때까지 진행.냄새도 없어짐.
            else:
                for x in range(cx+2,M):
                    # 이동못하면 이전칸에 멈추기.
                    if maze[cy][x]==0 or (maze[cy][x]==3 and o==0):
                        q.append((x-1,cy, 0))
                        break           
                    # 이동가능하고 nonpurple인경우 그 칸에 멈추기
                    elif maze[cy][x]==1 or maze[cy][x]==2 or (maze[cy][x]==3 and o==1) or x==M-1:
                        q.append((x,cy,0))
                        break
                        
        # 3. 위로 가는경우.. 이경우 답 안나옴.
        if cy>0 and not visit[cy-1][cx]:
            # red로 못감
            if maze[cy-1][cx]==0:
                visit[cy-1][cx] = True
            # pink는 그대로 진행
            elif maze[cy-1][cx]==1:
                visit[cy-1][cx] = True
                q.append((cx,cy-1,o))
            # orange인경우 오렌지냄새가 생김! 방문체크 해줘야하나...
            elif maze[cy-1][cx]==2:
                visit[cy-1][cx] = True
                q.append((cx,cy-1,1))
            # blue인경우 오렌지향이 있어야만 가능
            elif maze[cy-1][cx]==3 and o==1:
                visit[cy-1][cx] = True
                q.append((cx,cy-1,1))
            # !!!!! purple인경우. 그 방향으로 nonpurple이 나오거나, 못가는 칸이 나올때까지 진행.냄새도 없어짐.
            else:
                for y in range(cy-2,-1,-1):
                    # 이동못하면 이전칸에 멈추기.
                    if maze[y][cx]==0 or (maze[y][cx]==3 and o==0):
                        q.append((cx,y+1, 0))
                        break                
                    # 이동가능하고 nonpurple인경우 그 칸에 멈추기
                    elif maze[cy][x]==1 or maze[cy][x]==2 or (maze[cy][x]==3 and o==1) or y==0:
                        q.append((x,cy,0))
                        break      
                    
        # 4. 아래로 가는 경우! 답 가능
        if cy<N-1 and not visit[cy+1][cx]:
            # 답인경우!
            if (cx, cy+1)==(M-1,N-1):
                print(move)
                exit()
            # red로 못감
            if maze[cy+1][cx]==0:
                visit[cy+1][cx] = True
            # pink는 그대로 진행
            elif maze[cy+1][cx]==1:
                visit[cy+1][cx] = True
                q.append((cx,cy+1,o))
            # orange인경우 오렌지냄새가 생김! 방문체크 해줘야하나...
            elif maze[cy+1][cx]==2:
                visit[cy+1][cx] = True
                q.append((cx+1,cy,1))
            # blue인경우 오렌지향이 있어야만 가능
            elif maze[cy+1][cx]==3 and o==1:
                visit[cy+1][cx] = True
                q.append((cx,cy+1,1))
            # !!!!! purple인경우. 그 방향으로 nonpurple이 나오거나, 못가는 칸이 나올때까지 진행.냄새도 없어짐.
            else:
                for y in range(cy+2,N):
                    # 이동못하면 이전칸에 멈추기.
                    if maze[y][cx]==0 or (maze[y][cx]==3 and o==0):
                        q.append((cx,y-1, 0))
                        break                           
                    # 이동가능하고 nonpurple인경우 그 칸에 멈추기
                    elif maze[y][cx]==1 or maze[y][cx]==2 or (maze[y][cx]==3 and o==1) or y==N-1:
                        q.append((cx,y,0))
                        break
    move += 1
```
