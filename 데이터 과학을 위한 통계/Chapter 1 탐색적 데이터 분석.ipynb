확률론: 통계학의 수학적 기반. 순수 이론적 측면.  
통계학: 데이터 분석과 모델링을 다루는 일종의 **응용과학**  
탐색적 데이터 분석(Exploratory Data Analysis; EDA)는 통계학에서 비교적 새로운 영역이다.  
이전 통계학에서는 **추론**, 즉 적은 표본(샘플)을 가지고 더 큰 모집단에 대한 결론을 도출하기 위한 과정에 관해 다루었다.  

## 1.1 정형화된 데이터의 요소
데이터 과학에서 가장 중요한 도전은 가공되지 않은 원시 데이터(raw data)를 활용 가능한 형태의 정보로 변환하는 것.  
**용어 정리**  
- 연속형(continuous): 일정 범위 안에서 어떤 값이든 취할 수 있는 데이터(구간형, 실수형, 수치형 데이터)
- 이산(discrete): 횟수와 같은 정수 값만 취할 수 있다.(정수형, 횟수 데이터)
- 범주형(categorical): 가능한 범주 안의 값만을 취할 수 있다.(목록, 열거, 요인, 명목, 다항형 데이터)
- 이진(binary): 두 개의 값(0/1 혹은 참/거짓)만을 갖는 범주형 데이터의 특수한 경우.(이항적, 논리형, 지표, 불리언 데이터)
- 순서형(ordinal): 값들 사이에 분명한 순위가 있는 범주형 데이터.(정렬된 요인 데이터)

데이터를 분석하고 예측을 모델링할 때, 시각화, 해석, 통계 모델 결정 등에 데이터 종류가 중요한 역할을 한다.

## 1.2 테이블 데이터(Rectangular Data)
엑셀 스프레드시트나 데이터베이스의 테이블과 같은 테이블 데이터가 데이터 분석에서 가장 대표적으로 사용되는 객체의 형태이다.  
**용어 정리**
- 데이터 프레임(Data Frame): 통계와 머신러닝 모델에서 가장 기본이 되는 테이블 형태의 데이터 구조.
- 피처(Feature): 테이블의 각 열이 하나의 피처를 의미(특징, 속성, 입력, 예측변수, 변수)
- 결과(Outcome): 데이터 과학의 목표는 대부분 어떤 결과를 예측하는 것. 결과를 예측하기 위해 피처를 사용(종속변수, 응답, 목표, 출력)
- 레코드(Record): 테이블의 각 행은 하나의 레코드를 의미(기록값, 사건, 사례, 예제, 관측값, 패턴, 샘플)

데이터가 항상 이런 형태로 얻어지지는 않으므로 문자열과 같은 비정형 데이터는 테이블 데이터의 피처 형태로 표현되도록 처리해야 한다.  

### 1.2.1 데이터 프레임과 인덱스
데이터베이스에서 하나 혹은 그 이상의 열을 인덱스로 지정하여 SQL 쿼리 성능을 크게 향상할 수 있다.  

### 1.2.2 테이블 형식이 아닌 데이터 구조
- 시계열 데이터는 동일한 변수 안에 연속적인 측정값을 갖는다.  
- 지도 제작과 위치 정보 분석에 사용되는 공간 데이터의 경우 테이블 데이터보다 좀 더 복잡하고 다양하다.  
    - 어떤 객체(주택과 같은..)를 표현할 때는 객체와 그것의 공간 좌표가 데이터의 중심이 된다.
    - 필드정보는 공간을 나타내는 작은 단위들과 적당한 측정 기준값(픽셀의 밝기)에 중점을 둔다.
- 그래프(네트워크) 데이터는 물리적 관계, 사회적 관계, 추상적인 관계들을 표현하기 위해 사용된다.  

## 1.3 위치 추정

데이터의 변수들은 각기 다른 값을 갖는다. 데이터를 살펴보는 가장 기초적인 단계는 각 피처(변수)의 대푯값을 구하는 것.  
대부분의 값이 어디쯤에 위치하는지(중심경향성)을 나타내는 추정값이다.  
**용어 정리**
- 평균(mean): 모든 값의 총합을 개수로 나눈 값(average)
- 가중평균(weighted mean): 가중치를 곱한 값의 총합을 가중치의 총합으로 나눈 값
- 중간값(median): 데이터에서 가장 가운데 위치한 값(50번째 백분위수)
- 가중 중간값: 데이터를 정렬한 후, 각 가중치 값을 위에서부터 더할 때, 총합의 중간이 위치하는 데이터 값
- 절사평균(Trimmed Mean): 정해진 개수의 극단값(Extreme Value)을 제외한 나머지 값들의 평균(절단평균; truncated mean)
- 로버스트(Robust)하다: 극단값들에 민감하지 않다는 것을 의미(저항성이 있다;Resistant)
- 특잇값(Outlier): 대부분의 값과 매우 다른 데이터 값(극단값)

### 1.3.1 평균

가장 기본적인 위치 추정 방법.  
조금 변형한 것으로 **절사평균**이 있다.  
값들을 크기 순으로 정렬한 후, 양끝에서 일정 개수의 값들을 삭제한 뒤 남은 값들을 가지고 평균을 구한 것.  
절사평균은 극단값의 영향을 제거한다.  
또한 각 데이터 값 $x_i$에 가중치 $w_i$를 곱한 값들의 총합을 다시 가중치의 총합으로 나눈 **가중평균**이 있다.  
어떤 값들이 본래 다른 값들에 비해 큰 변화량을 갖을 때, 이러한 관측값에 대해 더 작은 가중치를 줄 수 있다.  
또 데이터를 수집할 때 우리가 관심있는 서로 다른 대조군에 대해서 항상 똑같은 수가 얻어지지는 않는다.  
이를 보정하기 위해서, 데이터가 부족한 소수 그룹에 대해 더 높은 가중치를 적용할 필요도 있기 때문에 가중평균을 사용하게 되었다.  

### 1.3.2 중간값과 로버스트 추정
데이터를 일렬로 정렬했을 때, 한가운데에 위치하는 값을 중간값이라고 한다. 데이터 개수가 짝수면 가운데 두 값의 평균.  
중간값이 많은 경우 위치 추정에 더 유리함. 가중 중간값도 있다.  
**특잇값**  
중간값은 결과를 왜곡할 수도 있는 특잇값(극단값)의 영향을 받지 않으므로 **로버스트** 위치 추정 방법으로 알려져 있다.  
특잇값은 어떤 데이터 집합에서 다른 값들과 매우 멀리 떨어져 있는 값들을 말한다.  
관습적인 정의가 있긴 하지만 정확한 정의는 다소 주관적일 수 있다.  
특잇값은 데이터 값 자체가 유효하지 않다거나 잘못되었다는 뜻이 아니다.(맞을 수도 있지만.. 에러값들일 수도.)  
특잇값들이 에러값이라면 평균은 잘못된 위치 추정을 할 수 있지만 중간값을 여전히 설득력 있는 결과를 줄 것.  
**이상 검출(Anomaly Detection)**  
이상 검출에서는 대부분의 정상적인 데이터보다는 예외적으로 측정된 특잇값들이 주된 관심의 대상이 된다.  
중간값만이 로버스트한 위치 추정방법은 아니고, 절사평균 역시 중간값과 평균의 절충안으로 많이 사용됨.

## 1.4 변이 추정
데이터의 특징을 요약하는 두 번째 요소인 **변이(Variability)**는 데이터 값이 얼마나 밀집해 있는지 혹은 퍼져 있는지를 나타내는 **산포도(Dispersion)**를 나타낸다.  
통계의 핵심이 바로 이 변이이다.  
**용어정리**  
- 편차(Deviation): 관측값과 위치 추정값사이의 차이(잔차, 오차)
- 분산(Variance): 평균과의 편차를 제곱한 값들의 합을 n-1로 나눈 값, n은 데이터 개수(평균 제곱오차)
- 표준편차(Standard Deviation): 분산의 제곱근(l2 Norm, 유클리드 노름)
- 평균절대편차(Mean Absolute Deviation): 평균과의 편차의 절댓값의 평균(l1 Norm, 맨하탄 노름)
- 중간값의 중위절대편차(Median Absolute Deviation from the Median): 중간값과의 편차의 절댓값의 중간값
- 범위(Range): 데이터의 최댓값과 최솟값의 차이
- 순서통계량(Order Statistics): 최소에서 최대까지 정렬된 데이터 값에 따른 계량형(순위)
- 백분위수(Percentile): 어떤 값들의 P퍼센트가 이 값 혹은 더 작은 값을 갖고, (100-P)퍼센트가 이 값 혹은 더 큰 값을 갖도록 하는 값(분위수)
- 사분위범위(Interquartile Range): 75번째 백분위수와 25번째 백분위수 사이의 차이(IQR)

### 1.4.1 표준편차와 관련 추정값들
가장 대표적으로 사용되는 변위 추정들은 관측 데이터와 위치 추정값 사이의 차이, 즉 **편차**를 기본으로 함.  
음의 편차는 양의 편차를 상쇄하기 때문에 편차 자체의 평균을 구하는 것은 바람직하지 않고, 편차의 절댓값을 구할 순 있음. **평균절대편차**  
가장 유명한 변이 추정 방법은 제곱 편차를 이요하는 **분산**과 **표준편차**  
분산은 제곱 편차의 평균이고, 표준편차는 분산의 제곱근이다.  
표준편차는 원래 데이터와 같은 척도(scale)에 있기 때문에 해석하기 쉽다.  
절댓값보다 제곱한 값이 수학적으로 다루는 데 더 편리하기 때문에 주로 사용.  

**자유도 n 아니면 n-1?**  
표준편차는 표본의 평균에 따른다는 **하나의 제약 조건**을 가지고 있기 때문에 n-1의 자유도(Degree of Freedom)을 갖는다.  
따라서 분산 수식에 n을 분모로 사용하면 모집단의 분산과 표준편차의 참값을 과소평가하게 된다. 이것이 **편향(biased)추정**  
n-1로 제대로 나누면 **비편향(Unbiased)추정**  

분산, 표준편차, 평균절대편차 모두 특잇값과 극단값에 로버스트하지 않다.  
로버스트한 변위 추정값으로는 중간값으로부터의 **중위절대편차(MAD)**가 있다.  
혹은 절사평균과 유사하게 절사 표준편차를 계산하는 것 역시 가능하다.  

### 1.4.2 백분위수에 기초한 추정
변위를 추정하는 또 다른 접근은 정렬된 데이터가 얼마나 퍼져 있는지를 보는 것이다.  
정렬 데이터를 나타내는 통계량을 **순서통계량**이라고 부른다.  
가장 기본이 되는 측도는 가장 큰 값과 작은 값의 차이를 나타내는 **범위**  
최대 최솟값 자체가 특잇값을 분석하는 데 큰 도움을 준다. 범위는 특잇값에 매우 민감.  
특잇값에 민감한 것을 피하기 위해, 범위의 양 끝에서 값들을 지운 후, 범위를 다시 알아 볼 수 있다.  
**백분위수** 사이의 차이를 가지고 이런 식의 추정을 할 수 있음.  
변위를 측정하는 가장 대표적인 방법은 **사분위범위(IQR)**라는, 25번째 백분위수와 75번째 백분위수의 차이를 보는 것.  

### 1.4.3 예제: 주별 인구의 변위 추정

```r
sd(state[["Population"]])  # 표준 편차
```

```r
IQR(state[["Population"]]) # 사분위범위
```

```r
mad(state[["Population"]]) # 중위절대편차
```

표준편차는 특잇값에 민감하기 때문에 다른 값과 차이가 많이 남.

## 1.5 데이터 분포 탐색하기
지금까지 알아본 추정들은 데이터의 위치 혹은 변이를 나타내기 위한 하나의 수치로 데이터를 요약.  
이와 더불어 데이터가 전반적으로 어떻게 분포하고 있는지를 알아보는 것 역시 유용하다.  
**용어 정리**  
- 상자그림(Boxplot): 투키가 데이터의 분포를 시각화하기 위한 간단한 방법으로 소개한 그림(상자수염도)
- 도수분포표(Frequency Table): 어떤 구간(Interval)(빈 bin)에 해당하는 수치 데이터 값들의 빈도를 나타내는 기록
- 히스토그램(Histogram): x축은 구간들을, y축은 빈도수를 나타내는 도수 테이블의 그림
- 밀도 그림(Density Plot): 히스토그램을 부드러운 곡선으로 나타낸 그림. 커널밀도추정(Kernel Density Estimation)을 주로 사용

### 1.5.1 백분위수와 상자그림

```r
quantile(state[["Murder.Rate"]], p=c(.05, .25, .5, .75, .95))
```

```r
boxplot(state[["Population"]]/1000000, ylab="Population (millions)")
```

상자부분의 위쪽과 아래쪽은 각각 75%, 25% 백분위수를 나타냄.  
중간값을 상자 안에 있는 굵은 수평선.  
구렛나룻처럼 위아래로 나 있는 점선이 바로 수염(Whisker)으로 데이터 전체의 범위를 나타내주는 위 아래 선들과 연결되어 있다.  
기본 설정상, R 함수는 수염 부분이 사분위범위의 1.5배 이상 더 멀리 나가지 않도록 한다.  
수염 부분보다 더 바깥쪽에 위치한 데이터는 각자 하나의 점으로 표시한다.  

### 1.5.2 도수분포표와 히스토그램
도수분포표는 변수의 범위를 **동일한 크기의 구간**으로 나누고 각 구간마다 몇개의 변수 값이 존재하는지를 보여줌

```r
breaks <- seq(from=min(state[["Population"]]),
             to=max(state[["Population"]]), length=11)
pop_freq <- cut(state[["Population"]], breaks=breaks,
               right=TRUE, include.lowest=TRUE)
table(pop_freq)
```

빈 구간이 있다는 것이 중요한 정보를 준다.

보통 이러한 모멘트값들을 직접 구하기보다는 시각화해서 직접 확인한다.

## 1.7 상관관계
모델링 프로젝트에서 EDA라 하면 **예측값들 간의 혹은 예측값과 목푯값과의 상관관계**를 조사해보는 것이 아주 중요.  
X가 큰 값을 가지면 Y도 큰 값을 갖고, X가 작은 값을 가지면 Y도 작은 값을 갖는 경우, 변수 X와 Y(측정된 값들)는 서오 양의 상관관계를 갖는다고 말함.  
반대의 경우면 음의 상관관계  
**용어정리**  
- 상관계수(Correlation Coefficient): 수치적 변수들 간에 어떤 관계가 있는지를 나타내기 위해 사용되는 측정량(-1 ~ 1 사이 값)
- 상관행렬(Correlation Matrix): 행과 열이 변수들을 의미하는 표. 각 셀은 그 행과 열에 해당하는 변수들 간의 상관관계를 의미
- 산점도(Scatterplot): x축과 y축이 서로 다른 두 개의 변수를 나타내는 도표

**상관계수(피어슨 상관계수)**라는 표준화된 방식으로 상관관계를 나타냄.  
두 변수 사이의 상관관계를 항상 같은 척도에 놓고 추정하는 것.  
변수 1과 변수 2 각각의 평균으로부터의 편차들을 서로 곱한 값들의 평균을 각 변수의 표준편차의 곱으로 나눠준다.  
  
$$r = {\sigma_{i=1}^N(x_i-\bar{x})(y_i-\bar{y})\over (N-1)s_xs_y}$$

N 대신 N-1로 나눠주는것에 유의..(자유도) 상관계수는 항상 +1(완전한 양의 상관관계)와 -1(완전한 음의 상관관계) 사이에 존재.  
변수들이 선형적인 관계를 갖지 않을 경우 상관계수는 더 이상 유용한 측정 지표가 아니다.

```r
## Code for CorrTable (Table 1.7)
library(ascii)
sp500_px <- read.csv(file.path('./psds_data/sp500_px.csv'))
sp500_sym <- read.csv(file.path('./psds_data/sp500_sym.csv'), stringsAsFactors = FALSE)
telecom <- sp500_px[, sp500_sym[sp500_sym$sector=="telecommunications_services", 'symbol']]
telecom <- telecom[row.names(telecom)>"2012-07-01", ]
telecom_cor <- cor(telecom)
ascii(telecom_cor, digits=c( 3,3,3,3,3), align=c("l", "r", "r", "r", "r", "r"), caption="Correlation between telecommunication stock returns.",
      include.rownames = TRUE, include.colnames = TRUE)
```

상관관계를 나타내는 상관행렬을 이용해 여러 가지 변수들 사이의 관계를 시각화하는데 사용한다.  
행렬의 대각 원소들은 모두1이고, 대각원소 아래쪽은 위쪽과 같은 대칭행렬이다.  

```r
sp500_px <- read.csv(file.path('./psds_data/sp500_px.csv'))
sp500_sym <- read.csv(file.path('./psds_data/sp500_sym.csv'), stringsAsFactors = FALSE)
etfs <- sp500_px[row.names(sp500_px)>"2012-07-01",
                sp500_sym[sp500_sym$sector=="etf", 'symbol']]
library(corrplot)
corrplot(cor(etfs), method="ellipse")
```

주요 상장 주식 펀드(ETF)들의 일간 수익 사이의 상관관계.  
ETF는 S&P 500(SPY)와 다우 존스 지수(DIA)와 높은 상관성을 갖는다.  
타원 모양의 방향은 두 변수가 양의 관계를 갖는지(타원이 오른쪽으로 기울어져 있음)  
아니면 음의 관계를 갖는지(타원이 왼쪽으로 기울어져 있음)를 나타냄.  
타원의 색깔과 너비는 상관관계의 강도를 의미. 얇고 진할수록 더 강한 관계성을 나타낸다.  
평균과 표준편차와 같이, 상관계수는 데이터의 특잇값에 민감하다.  

**상관관계를 추정하는 다른 방법**  
스피어먼의 로(Spearman's rho)나 켄들의 타우(Kendall's tau)와 같은 다른 형태의 상관계수들이 있다.  
이들은 데이터의 순위를 기초로 하는 상관계수값.  
값 자체보다 값의 순위를 이용하기 때문에 특잇값에 좀 더 로버스트하며 비선형 관계도 다룰 수 있다.  
데이터과학에서는 그냥 피어슨 상관계수, 혹은 이것의 로버스트한 다른 버전들을 탐색 분석에 주로 사용함.  
순위 기반 추정법은 보통 데이터의 크기가 작고 어떤 특별한 가설검정이 필요할 때 주로 사용된다.

## 1.8 두 개 이상의 변수 탐색하기
평균과 분산과 같이 익숙한 추정값들은 한 번에 하나의 변수를 다룬다.(**일변량분석(Univariate Analysis)**)  
상관분석은 두 변수(**이변량분석(Bivariate Analysis)**)를 비교할 때 중요한 방법.  
이번 절에서는 이에 관한 추정법과 도표를 살펴보고 셋 이상의 변수(**다변량분석(Multivariate Analysis)**)를 다룸.  
**용어 정리**  
- 분할표(Contingency Table): 두 가지 이상의 범주형 변수의 빈도수를 기록한 표
- 육각형 구간(Hexagonal Binning): 두 변수를 육각형 모양의 구간으로 나눈 그림
- 등고 도표(Contour Plot): 지도상에 같은 높이의 지점을 등고선으로 나타내는 것처럼, 두 변수의 밀도를 등고선으로 표시한 도표
- 바이올린 도표(Violin Plot): 상자그림과 비슷하지만 밀도추정을 함께 보여준다.

일변량분석과 마찬가지로, 이변량분석 역시 요약통계를 계산하고 시각화하는 것을 기본으로 한다.  
이변량분석 혹은 다변량분석의 형태는 데이터가 수치형인지 아니면 범주형인지, 데이터의 특성에 따라 달라진다.

### 1.8.3 범주형 변수 대 수치형 변수
상자그림은 범주형 변수에 따라 분류된 수치형 변수의 분포를 시각화하여 비교하는 간단한 방법이다.  

```r
airline_stats <- read.csv(file.path('./psds_data/airline_stats.csv'), stringsAsFactors = FALSE)
airline_stats$airline <- ordered(airline_stats$airline, levels=c('Alaska', 'American', 'Jet Blue', 'Delta', 'United', 'Southwest'))
boxplot(pct_delay ~ airline, data=airline_stats, ylim=c(0, 50))
```

위 그림은 한 달 동안 항공기와 관련하여 일어난 비행 지연의 비율을 보여준다.  
알래스카 항공의 지연이 가장 적었던 반면, 아메리카 항공의 항공 지연이 가장 많았다.  
아메리카 항공의 낮은 사분위수가 알래스카 항공의 상위 사분위수보다도 높았다.  

**바이올린 도표**는 상자그림을 보완한 형태로, y축을 따라 밀도추정 결과를 동시에 시각화한다.  
밀도 분포 모양을 좌우대칭으로 서로 겹쳐지도록 하여 보면 바이올린을 닮은 모양.  
바이올린 도표의 장점은 상자그림에서는 보이지 않는 데이터의 분포를 볼 수 있다는 점이다.  
한편 상자그림은 데이터의 특잇값들을 좀 더 명확하게 보여준다.

```r
ggplot(data=airline_stats, aes(airline, pct_carrier_delay)) +
ylim(0, 50) +
geom_violin() +
labs(x="", y="Daily % of Delayed Flights")
```

이 바이올린 도표를 통해 알래스카 항공, 델타 항공이 거의 0근처에 데이터가 집중되어 있는 것을 볼 수 있다.  
상자그림에서는 명확하게 드러나지 않는 현상이었다.    
geom_boxplot 함수를 추가한다면 바이올린 도표에 상자그림을 결합할 수도 있다.
